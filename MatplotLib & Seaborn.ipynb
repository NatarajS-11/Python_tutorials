# MatPlotLib is library for data Visualization

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
y= x*2
y1 =x*3
```

```python
#Line plot
plt.plot(x,y,color='red',linestyle=":",linewidth=2)
plt.title("Line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#Line plot
plt.plot(x,y,color='red',linestyle=":",linewidth=2)
plt.plot(x,y1,color='green',linestyle="-",linewidth=2)
plt.title("Line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#subplot
plt.subplot(1,2,1)
plt.plot(x,y,color='red',linestyle=":",linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y1,color='green',linestyle="-",linewidth=2)
plt.title("Line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#bar graph
d= {"bob":67,"Matt":88,"sam":90}
d
```

```python
type(d)
```

```python
Names=list(d.keys())
Marks= list(d.values())
```

```python
#vertical bar graph
plt.bar(Names,Marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#horizontal bar graph
plt.barh(Names,Marks, color='g')
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#scatterplot to understand the distribution between two values 
x = [1,25,3,46,5,69,7,8]
y = [10,20,30,40,50,60,70,80]
z =[98,76,98,54,12,45,83,76]
plt.scatter(x,y,marker="*", s=100)
plt.scatter(x,z,marker=".",c='orange', s=100)
plt.show()
```

```python
#subplot for scatterplot
x = [1,25,3,46,5,69,7,8]
y = [10,20,30,40,50,60,70,80]
z =[98,76,98,54,12,45,83,76]
plt.subplot(1,2,1)
plt.scatter(x,y,marker="*", s=100)
plt.subplot(1,2,2)
plt.scatter(x,z,marker=".",c='orange', s=100)
plt.show()
```

```python
#histogram 
x = [1,1,1,1,1,2,2,2,2,4,5,6,7,8,9,9,9] #here 1 is occuring 5 times
plt.hist(x,bins=8,color="r")
plt.show()
```

```python
#boxplot 

one = [1,2,3,4,5,6,7,8]
two = [10,20,30,40,50,60,70,80]
three =[45,6,7,8,9,1,2,9]
data = list([one,two,three])
```

```python
plt.boxplot(data)
plt.show()
```

```python
#violin chart
plt.violinplot(data,showmedians=True)
plt.show()
```

```python
#PieChart
fruits= ["orange","banana","apple"]
quantity =[200,100,500]
plt.pie(quantity,labels=(fruits),autopct='%01.f%%')
plt.show()
```

```python
#donut chart
plt.pie(quantity,labels=(fruits),radius=1)
plt.pie([5],colors= ['w'],radius=0.5)
plt.show()
```

# Seaborn 

```python
import seaborn as sns
```

```python
ds = sns.load_dataset("fmri")
ds.head()
```

```python
#Lineplot
sns.lineplot(x="timepoint",y="signal",hue="event",data=ds,style="event",markers=True)
plt.show()
```

```python
df = pd.read_csv('pokemon.csv')
df.head()
```

```python
sns.barplot(x='is_legendary',y="weight_kg",hue="generation", data=df, palette='rocket')
plt.show()
```

```python
sns.barplot(x='is_legendary',y="weight_kg", data=df, palette='Blues_d')
plt.show()
```

```python
#Scatterplot
df2 = pd.read_csv("Iris.csv")
df2
```

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=df2,hue='PetalLengthCm')
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=df2,hue='Species',style='Species')
plt.show()
```

```python
#histogram/distplot
df3= pd.read_csv('diamonds.csv')
df3.head()
```

```python
sns.distplot(df3['price'])
plt.show()
```

```python
sns.distplot(df3['price'],color='green')
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',data=df2, color='orange') #is a combination of histogram and scatterplot
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',data=df2, color='orange',hue='Species') #is a combination of histogram and scatterplot
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',data=df2, color='olive',kind='reg') #is a combination of histogram and scatterplot
plt.show()
```
